```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.stats import mvn
%matplotlib inline  
```

```python
mu = np.array([0,0])
mu2 = np.array([2,2])
Sigma = np.array([[.25, .3],[.3, 1]])
x1 = np.linspace(-3,5,(5-(-3))/.2)
x2 = np.linspace(-6,7,(7-(-6))/.2)

[X1, X2] = np.meshgrid(x1,x2);
X1 = X1.reshape(np.size(X1),1);
X2 = X2.reshape(np.size(X2),1);
X = np.column_stack((X1, X2));

F = multivariate_normal.pdf(X, mu,Sigma)
F2 = multivariate_normal.pdf(X, mu2,Sigma)

F = F.reshape(np.size(x2), np.size(x1))
F2 = F2.reshape(np.size(x2), np.size(x1))

mvn.mvnun(np.array([0,0]), np.array([1,1]), mu, Sigma)

level = [.0001, .001, .01]
for i in range(9):
    level.append(0.05+0.1*i)
level.append(.99)
level.append(.999)
level.append(.9999)

[X1, X2] = np.meshgrid(x1,x2);

plt.figure(1, figsize=(7,5))
plt.contour(X1, X2, F, level)
plt.contour(X1, X2, F2, level)
plt.xlabel('x1')
plt.xlabel('x2')
plt.show()
```
