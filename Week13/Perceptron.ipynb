```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib inline
```

```python
Xpoints = np.arange(-100, 110, 10)
numPoints = len(Xpoints)
X1 = np.zeros((numPoints, numPoints))
X2 = np.zeros((numPoints, numPoints))
Y_sigmoid1 = np.zeros((numPoints, numPoints))
```

```python
cnt1 = 0
cnt2 = 0
for i in Xpoints:
    cnt2 = 0
    for j in Xpoints:
        X1[cnt1, cnt2] = i
        X2[cnt1, cnt2] = j        
        cnt2 += 1
    cnt1 += 1
```

```python
A = np.array([0.5, 0.5])
b1 = -50
b2 = 50
```

```python
Y_sigmoid1 = 1/(1+np.exp(-(X1*A[0] + X2*A[1] + b1)))
Y_sigmoid2 = 1/(1+np.exp(X1*A[0] + X2*A[1] + b2))
Y_sigmoid3 = 1/(1+np.exp(-(100*Y_sigmoid1 + 100*Y_sigmoid2 - 100)))
```

```python
fig = plt.figure(figsize = (19, 6))
ax = fig.add_subplot(1, 3, 1, projection='3d')
ax.plot_surface(X1, X2, Y_sigmoid1, rstride = 1, cstride = 1, cmap = 'PuBu')
ax.text2D(0.05, 0.95, "Y_sigmoid1", transform=ax.transAxes)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y_sigmoid1')

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.plot_surface(X1, X2, Y_sigmoid2, rstride = 1, cstride = 1, cmap = 'PuBu')
ax.text2D(0.05, 0.95, "Y_sigmoid2", transform=ax.transAxes)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y_sigmoid2')

ax = fig.add_subplot(1, 3, 3, projection='3d')
ax.plot_surface(X1, X2, Y_sigmoid3, rstride = 1, cstride = 1, cmap = 'PuBu')
ax.text2D(0.05, 0.95, "Y_sigmoid3", transform=ax.transAxes)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y_sigmoid3')

plt.show()
```
