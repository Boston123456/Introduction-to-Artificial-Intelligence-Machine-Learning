```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(-2, 2, 0.01)

logloss = np.log(1 + np.exp(-x + 0.5))
hingeloss = np.zeros(x.shape[0])
zerooneloss = np.zeros(x.shape[0])
for i in range(x.shape[0]):
    hingeloss[i] = max(0, 1 - x[i])
    zerooneloss[i] = max(0, np.sign(-x[i]))
```

```python
plt.figure(1, figsize = (8, 7))
plt.plot(x, logloss, 'b-', linewidth = 4, label = 'log loss')
plt.plot(x, hingeloss, 'r-', linewidth = 4, label = 'hinge loss')
plt.plot(x, zerooneloss, 'k-', linewidth = 4, label = 'zero one loss')

plt.xlabel('x', fontsize = 15)
plt.legend()
plt.show()
```
