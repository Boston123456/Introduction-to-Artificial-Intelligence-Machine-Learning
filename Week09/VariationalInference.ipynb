```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline  
```

```python
x = np.arange(-10,10,0.01);
N = np.size(x) - 1;
y = np.divide(np.ones((1,np.size(x))), (1+np.exp(-x)))[0];

plt.figure(1, figsize=(7,5));
p1 = plt.plot(x,y, color="red");

for randa in np.arange(0,10,0.1):
    H = -randa * np.log(randa) - (1-randa) * np.log(1-randa);
    y = np.exp(randa * x - H);

randa = 0.2
H = -randa * np.log(randa) - (1-randa) * np.log(1-randa);
y = np.exp(randa * x - H);
p2 = plt.plot(x,y, color="blue");

randa = 0.7
H = -randa * np.log(randa) - (1-randa) * np.log(1-randa);
y = np.exp(randa * x - H);
p3 = plt.plot(x,y, color="black");

plt.xlim((-10,10));
plt.ylim((0,1));
plt.show();
```
