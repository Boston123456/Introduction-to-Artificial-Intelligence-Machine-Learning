```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

%matplotlib inline  
```

```python
x = np.arange(0.01,3,0.01)
y_exp = np.exp(x[:])
y_tal_exp = np.zeros((np.size(x),1))

for i in range(np.size(x)):
    y_tal_exp[i] = 1;
    for j in range(2):
        y_tal_exp[i] = y_tal_exp[i] + 1/factorial(j+1) * np.power((x[i] - 0),j+1)

plt.figure(1, figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x,y_exp, color='r');
plt.plot(x,y_tal_exp, color='b')

y_log = np.log(x[:])
y_tal_log = np.zeros((np.size(x), 1))
a = 0.5

for i in range(np.size(x)):
    y_tal_log[i] = np.log(a)
    for j in range(8):
        y_tal_log[i] = y_tal_log[i] + np.power(-1, j+2) * np.power(1/a, j+1) / factorial(j+1) * np.power(x[i]-a, j+1)

plt.subplot(1,2,2)
plt.figure(1, figsize=(7,5))
plt.plot(x,y_log, color='r')
plt.plot(x,y_tal_log, color='b')

plt.show()
```
